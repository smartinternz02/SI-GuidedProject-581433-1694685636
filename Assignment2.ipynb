# Take car crashes dataset from seaborn library and perform Data visualization steps

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('car_crashes')
```

```python
df
```

```python
sns.__version__
```

```python
df.info()
```

# 1.Scatter plot


```python
sns.scatterplot(x="alcohol",y="total",data=df)
```

```python
Inference: The scatterplot helps visualize the relationship between alcohol involvement and total crashes.
```

# 2.Histplot

```python
sns.histplot(x="total",bins=20, kde=True,data=df)
```

```python
Inference:The histplot allows you to draw the following inferences about the distribution of total crashes
```

# 3.Boxplot

```python
sns.boxplot(df)
```

```python
Inference: This boxplot helps visualize the relationship between all columns and their prices.
```

# 4.Countplot

```python
sns.countplot(x="abbrev",data=df)
```

```python
Inference:In this countplot we can see that relation between states and count of crashes occured in each state.
```

# 5.Jointplot

```python
sns.jointplot(x="alcohol",y="speeding",data=df)
```

```python
Inference:The joint plot above displays the relationship between the "alcohol" and "speeding" variables. 
```

# 6.Relplot

```python
sns.relplot(x="alcohol",y="speeding",data=df,height=6,hue="speeding")
```

```python
Inference:The relplot allows us to visually assess the relationship between alcohol involvement and speeding involvement in car crashes
```

# 7.Lineplot

```python
sns.lineplot(x="abbrev",y="total",data=df,marker="o")
```

```python
Inference:The plot visualizes how the total number of crashes varies from one state to another state.
```

# 8.Barplot

```python
sns.barplot(x="abbrev",y="total",data=df)
```

```python
Inference:The bar plot shows the average total crashes by state. 
```

# 9.Heatmap

```python
corr=df.corr()
corr 
```

```python
sns.heatmap(corr,annot=True)
```

```python
Inference:
1.Darker colors indicate stronger correlations, while lighter colors indicate weaker or no correlations.
2. The diagonal of the heatmap is always 1 because it represents the correlation of a variable with itself (perfect correlation).   
    
```
